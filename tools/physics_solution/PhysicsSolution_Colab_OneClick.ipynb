# Physics Solution Colab One-Click

MyDrive에 프로젝트 zip과 데이터 zip을 올린 뒤, 이 노트북에서 `런타임 > 모두 실행`만 누르면 전체 파이프라인이 돌아가도록 만든 실행용 노트북입니다.

- 프로젝트 zip 기본 경로: `/content/drive/MyDrive/physics_solution.zip`
- 데이터 zip 기본 경로: `/content/drive/MyDrive/open (7).zip`
- geometry reasoning 브랜치는 기본 OFF
- 학습 진행 상태는 Colab 출력창의 실시간 진행바로 확인

```python
from google.colab import drive
drive.mount('/content/drive')

from pathlib import Path

PROJECT_ZIP = Path('/content/drive/MyDrive/physics_solution.zip')
DATASET_ZIP = Path('/content/drive/MyDrive/open (7).zip')
LOCAL_PROJECT_ZIP = Path('/content/physics_solution_project.zip')
LOCAL_PROJECT_ROOT = Path('/content/physics_solution')
LOCAL_RUNTIME_ROOT = Path('/content/physics_solution_runtime')

assert PROJECT_ZIP.exists(), f'Project zip not found: {PROJECT_ZIP}'
assert DATASET_ZIP.exists(), f'Dataset zip not found: {DATASET_ZIP}'
print('Project zip:', PROJECT_ZIP)
print('Dataset zip:', DATASET_ZIP)
```

```python
import shutil

if LOCAL_PROJECT_ROOT.exists():
    shutil.rmtree(LOCAL_PROJECT_ROOT)
if LOCAL_PROJECT_ZIP.exists():
    LOCAL_PROJECT_ZIP.unlink()

shutil.copy2(PROJECT_ZIP, LOCAL_PROJECT_ZIP)
shutil.unpack_archive(str(LOCAL_PROJECT_ZIP), '/content')
print('Extracted project to:', LOCAL_PROJECT_ROOT)
```
